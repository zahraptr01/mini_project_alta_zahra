#### Penyesuaian tipe data untuk dataset projects_data.csv

```python
columns_to_drop = ['sector1', 'sector2', 'sector3', 'sector4', 'sector5',
                   'mjsector1', 'mjsector2', 'mjsector3', 'mjsector4', 'mjsector5', 'mjsector',
                   'theme1', 'theme2', 'theme3', 'theme4', 'theme5', 'goal', 'financier',
                   'mjtheme1name', 'mjtheme2name', 'mjtheme3name', 'mjtheme4name', 'mjtheme5name',
                   'location', 'GeoLocID', 'GeoLocName', 'Latitude', 'Longitude', 'Country', 'Unnamed: 56']

# Menghapus kolom-kolom yang tidak diperlukan
projects = projects.drop(columns=columns_to_drop)

print(projects)
```

#### Mengisi nilai yang hilang menggunakan string kosong(")

```python
projects_filled_strings = projects.fillna('')

# Menyimpan dataframe yang sudah diisi nilai
file_path_filled_strings = "../dataset/projects_data_filled_strings.csv"
projects_filled_strings.to_csv(file_path_filled_strings, index=False)
```
